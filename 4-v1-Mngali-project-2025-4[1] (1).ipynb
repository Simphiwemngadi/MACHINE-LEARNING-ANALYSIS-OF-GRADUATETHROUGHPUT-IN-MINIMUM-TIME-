```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_object(dataframe, column_name):
    """
    Plots a bar chart showing category frequencies with both frequency (inside bar)
    and proportion (above bar) labels.

    Parameters:
    - dataframe: pandas DataFrame
    - column_name: str, name of the categorical column to visualize
    """
    # Count frequencies and proportions
    value_counts = dataframe[column_name].value_counts()
    proportions = value_counts / len(dataframe)

    # Set plot style
    sns.set(style="whitegrid")
    plt.figure(figsize=(10, 6))

    # Bar plot
    #ax = sns.barplot(x=value_counts.index, y=value_counts.values, palette="muted")
    colors = sns.color_palette("Spectral", len(value_counts))
    ax = sns.barplot(x=value_counts.index, y=value_counts.values,palette=colors)

    # Annotate bars
    for i, (count, prop) in enumerate(zip(value_counts.values, proportions.values)):
        # Frequency inside bar
        ax.text(i, count * 0.5, f'{count}', ha='center', va='center',
                fontsize=11, color='black', fontweight='bold')
        # Proportion above bar
        ax.text(i, count + max(value_counts.values) * 0.02, f'{prop:.2%}',
                ha='center', fontsize=10, color='navy')

    plt.title(f'Distribution of {column_name}')
    plt.xlabel(column_name)
    plt.xticks(rotation=90)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

## Univariate analysis for continous variables

```python
def cont_plot(df, var):
    #var="Age"
    # Set plot style
    sns.set(style="whitegrid")

    # Create a figure with two subplots: histogram and box plot
    plt.figure(figsize=(12, 6))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[var], bins=20, kde=True, color='skyblue')
    plt.title(var+' Distribution - Histogram')
    plt.xlabel(var)
    plt.ylabel('Frequency')

    # Box plot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[var], color='lightgreen')
    plt.title(var+' Distribution - Box Plot')
    plt.xlabel(var)

    plt.tight_layout()
    plt.show()  
```

## Bivariate analysis between the Target variable and a continous variables

```python
def cat_cont_plot(df, xvar, yvar):
    """
    Creates histograms of a continuous variable across categories.
    
    Parameters:
    - df: pandas DataFrame
    - xvar: str, categorical column name (e.g. 'Target')
    - yvar: str, continuous column name (e.g. 'Age')
    """
    sns.boxplot(x=xvar, y=yvar, data=df)
    plt.xticks(rotation=90)
    
    
    # Create a larger facet grid of histograms by category
    g = sns.FacetGrid(df, col=xvar, col_wrap=4, height=6, aspect=0.5)  # Increased size and layout
    g.map(sns.histplot, yvar, bins=20, color="steelblue")

    # Add titles and labels
    g.set_axis_labels(yvar, "Frequency")
    g.set_titles(col_template="{col_name}")
    plt.tight_layout()
    plt.show()
```

## Bivariate analysis between the Target variable and a categorical variable

```python
def cat_cat_plot(dataframe, column_name, hue_column):
    """
    Plots a grouped bar chart showing category frequencies split by hue,
    with both frequency (inside bar) and proportion (above bar) labels.
    
    Parameters:
    - dataframe: pandas DataFrame
    - column_name: str, name of the categorical column to visualize (x-axis)
    - hue_column: str, name of the second categorical variable to group by (hue)
    """
    # Count combinations of column and hue
    counts_df = dataframe.groupby([column_name, hue_column]).size().reset_index(name='count')
    total_counts = dataframe[column_name].value_counts()

    # Set plot style
    sns.set(style="whitegrid")
    plt.figure(figsize=(14, 8))  # Larger frame

    # Create bar plot
    colors = sns.color_palette("Spectral")
    ax = sns.barplot(x=column_name, y='count', hue=hue_column, data=counts_df,palette=colors)
    #ax = sns.barplot(x=column_name, y='count', hue=hue_column, data=counts_df, palette="muted")
    # Annotate bars
    for container in ax.containers:
        for bar in container:
            height = bar.get_height()
            x = bar.get_x() + bar.get_width() / 2
            category = bar.get_label()
            base_x = int(round(x))  # used for proportion lookup
            if height > 0:
                ax.text(x, height * 0.5, f'{int(height)}', ha='center', va='center',
                        fontsize=10, color='black', fontweight='bold')
                ax.text(x, height + max(counts_df['count']) * 0.02, f'{height / sum(container.datavalues):.1%}',
                        ha='center', fontsize=10, color='navy', fontweight='bold')


    # Beautify plot
    plt.title(f'{column_name} Distribution by {hue_column}', fontsize=16)
    plt.xlabel(column_name, fontsize=23)
    plt.ylabel('Frequency', fontsize=19)
    plt.xticks(rotation=45)
    plt.legend(title=hue_column)
    plt.tight_layout()
    plt.show()

```

#### 1. Remove Constant and Quasi-Constant Features

```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold

# Step 1: Fit the selector
selector = VarianceThreshold(threshold=0.01)
selector.fit(df_BGENCV4_encoded)

# Step 2: Get retained and removed columns
selected_columns = df_BGENCV4_encoded.columns[selector.get_support()]
removed_columns = df_BGENCV4_encoded.columns[~selector.get_support()]

# Step 3: Transform the data
df_D4_enc_reduced1 = pd.DataFrame(selector.transform(df_BGENCV4_encoded), columns=selected_columns)

# Step 4: Display results
print(f"✅ Retained columns ({len(selected_columns)}):")
print(selected_columns.tolist())

print(f"\n❌ Removed columns ({len(removed_columns)}):")
print(removed_columns.tolist())

print(f"\n📐 New shape: {df_D4_enc_reduced1.shape}")
print(df_D4_enc_reduced1.head())
```

##### 2. Check for Perfect or Near-Perfect Correlation

```python
import numpy as np
import pandas as pd

# Step 1: Compute correlation matrix
corr_matrix = df_D4_enc_reduced1.corr().abs()

# Step 2: Extract upper triangle (to avoid duplicate pairs)
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Step 3: Identify columns to drop (correlation > 0.95)
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

# Step 4: Drop columns
df_D4_enc_reduced2 = df_D4_enc_reduced1.drop(columns=to_drop)

# Step 5: Display results
print(f"❌ Removed due to high correlation ({len(to_drop)}):")
print(to_drop)

print(f"\n✅ Remaining columns ({df_D4_enc_reduced2.shape[1]}):")
print(df_D4_enc_reduced2.columns.tolist())

print(f"\n📐 New shape: {df_D4_enc_reduced2.shape}")
print(df_D4_enc_reduced2.head())
```

```python
from sklearn.model_selection import train_test_split
# Define X and y from the original dataset
y = df_D4_enc_reduced2['Minimum Time']  # Target variable
X = df_D4_enc_reduced2.drop(columns=['Minimum Time'])  # Feature matrix

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=5, stratify=y  # stratify if classification
)
print("X train", X_train.shape)
print("y test", y_train.shape)
print("X test", X_test.shape)
print("y test", y_test.shape)
```

## Feature scaling: Apply log1p + MinMax scaling only on Numeric_cols

- np.log1p(x) is equivalent to log(x + 1) — it avoids issues with zero values. It reduces the impact of large values and spreads out smaller ones. After log transformation, the distribution becomes more balanced.
- Min-Max Scaling: This ensures: All values are between 0 and 1. The scale reflects relative differences more fairly. Outliers don’t distort the range

```python
#List of column removed when dropping constant and nearly constant columns
removed_cols=removed_columns.tolist()
removed_cols
```

```python
numeric_cols
```

```python
### Deducting nearly and constant features from numeric_cols list (original numeric column)

# Step 1: Find intersection
intersection = list(set(numeric_cols) & set(removed_cols))

# Step 2: Deduct intersection from numeric_cols
filtered_numeric_cols = [col for col in numeric_cols if col not in intersection]

# Step 3: Display results
print(f"🔗 Intersection (columns removed from numeric_cols): {intersection}")
print(f"📊 Updated numeric_cols ({len(filtered_numeric_cols)}): {filtered_numeric_cols}")
```

```python
# list of columns dropped from dropping highly correlated columns

to_drop 
```

```python
## deducting highly correlated columns from numeric_cols list (original numberic columns)

# Step 1: Find intersection
intersection2 = list(set(filtered_numeric_cols) & set(to_drop))

# Step 2: Deduct intersection from numeric_cols
filtered_numeric_cols2 = [col for col in filtered_numeric_cols if col not in intersection2]

# Step 3: Display results
print(f"🔗 Intersection (columns removed from numeric_cols): {intersection2}")
print(f"📊 Updated numeric_cols ({len(filtered_numeric_cols)}): {filtered_numeric_cols2}")
```

```python
from sklearn.preprocessing import MinMaxScaler
#import numpy as np


# Step 2: Apply log1p + MinMax scaling only to selected columns
scaler = MinMaxScaler()

for col in filtered_numeric_cols2:
    # Log transform
    train_log = np.log1p(X_train[col])
    test_log = np.log1p(X_test[col])
    
    # Fit scaler on training data
    train_scaled = scaler.fit_transform(train_log.values.reshape(-1, 1))
    test_scaled = scaler.transform(test_log.values.reshape(-1, 1))
    
    # Replace original columns with scaled values
    X_train[col] = train_scaled
    X_test[col] = test_scaled


```

## 3. feature selection using feature importance of xgboost

```python
# 3. feature selection using feature importance of xgboost
from xgboost import XGBClassifier
#import pandas as pd

model = XGBClassifier()
model.fit(X_train, y_train)

importance = pd.Series(model.feature_importances_, index=X.columns)
#This keeps only features with above-average importance, which balances pruning with performance.
threshold = importance.mean()  # You can also try 0.01 or importance.median()

selected_features = importance[importance > threshold].index
X_xgb_train = X_train[selected_features]
X_xgb_test = X_test[selected_features]
print(f"Selected {len(selected_features)} features out of {X.shape[1]}")
print("X_xgb_train", X_xgb_train.shape)
print("X_xgb_test", X_xgb_test.shape)
```
